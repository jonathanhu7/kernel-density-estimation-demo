```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import alpha
```

```python
plt.rcParams["figure.figsize"] = [12, 8]
x_range = np.linspace(-5, 5, 1000)
h_values = np.arange(0.2, 1.2, 0.2)
np.random.seed(42)
```

```python
def gaussian_kernel_function(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-(x ** 2) / 2)
```

```python
def plot_gaussian_kernel_function():
    y_range = gaussian_kernel_function(x_range)

    plt.plot(x_range, y_range, label="Gaussian Kernel")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Plot of Gaussian Kernel Function")
    plt.savefig("./img/plot-of-gaussian-kernel-function.png")

plot_gaussian_kernel_function()
```

```python
def plot_gaussian_kernel_functions_with_different_h_values():
    for h in h_values:
        y_range = gaussian_kernel_function(x_range / h)
        plt.plot(x_range, y_range, label=f"h = {h:.1f}")

    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Plot of Gaussian Kernel Function for Different h Values")
    plt.legend()
    plt.savefig("./img/plot-of-gaussian-kernel-function-for-different-h-values.png")
    
plot_gaussian_kernel_functions_with_different_h_values()
```

```python
def kernel_density_estimation():
    data = np.random.normal(size=30)
    plt.hist(data, bins=30, density=True, alpha=0.5, label="Data histogram")
    
    for h in h_values:
        pdf = np.zeros_like(x_range)
        
        for i, x in enumerate(x_range):
            kernel_values = gaussian_kernel_function((data - x) / h)
            pdf[i] = np.sum(kernel_values) / (h * len(data))
        
        plt.plot(x_range, pdf, label=f"h = {h:.1f}")
        
    plt.xlabel("x")
    plt.ylabel("density")
    plt.title("Gaussian Kernel Density Estimation")
    plt.legend()
    plt.savefig("./img/gaussian-kernel-density-estimation.png")
    
kernel_density_estimation()
```
